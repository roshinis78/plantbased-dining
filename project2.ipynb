```python
import pandas as pd
import numpy as np
import seaborn as sns

def extract_first_website(websiteList):
    split = websiteList.split(',', 1) # split string once by comma
    if split[0][0:4] == "http":
        return "https" + split[0][4:]
    return split[0]

df = pd.read_csv("restaurants.csv")
df = (df[df.country == "US"]).drop_duplicates(subset="id")
df = df[["city", "cuisines", "longitude", "latitude", "priceRangeMin", "priceRangeMax", "websites"]]
df = df.dropna(subset=["longitude", "latitude"])

df = df.fillna({'priceRangeMin':0,'priceRangeMax':0, 'websites':'', 'city':'Unknown', 'cuisines':''})
df['websites'] = (df['websites']).apply(extract_first_website)
df.to_csv("relevant_data.csv")
df
```
